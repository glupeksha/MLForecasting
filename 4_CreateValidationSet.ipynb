```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

#os - operating system dependent. to read/write files
#pd - pandas for data analysis/modeling

data_path = 'data/'

#get DataFrames
#the training set. Daily historical data from January 2013 to October 2015
trainingMonthSet    = pd.read_pickle(os.path.join(data_path, 'trainingMonthSet.pickle'))
```

```python
validationSet=trainingMonthSet.loc[trainingMonthSet.date_block_num==33]
validationSet.head()
```

```python
trainingSet=trainingMonthSet.loc[trainingMonthSet.date_block_num<33]
trainingSet.tail()
```

```python
trainingSetX=trainingMonthSet.copy().drop(['item_cnt_month','YearMonth'],axis=1)
trainingSetY=trainingMonthSet.copy().drop(['date_block_num','YearMonth','shop_id','item_id','item_category_id','item_price'],axis=1)
validationSetX=validationSet.copy().drop(['item_cnt_month','YearMonth'],axis=1)
validationSetY=validationSet.copy().drop(['date_block_num','YearMonth','shop_id','item_id','item_category_id','item_price'],axis=1)
```

```python
validationSetY.head()
```

```python
trainingSetX.to_csv(data_path+'trainingSetX.csv', index=False)
trainingSetY.to_csv(data_path+'trainingSetY.csv', index=False)
validationSetX.to_csv(data_path+'validationSetX.csv', index=False)
validationSetY.to_csv(data_path+'validationSetY.csv', index=False)
```
