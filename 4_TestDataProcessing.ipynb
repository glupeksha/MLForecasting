```python
import pandas as pd
import os
#os - operating system dependent. to read/write files
#pd - pandas for data analysis/modeling

data_path = 'data/'

items = pd.read_csv(os.path.join(data_path, 'items.csv'),encoding ='ISO-8859-1')

#get DataFrames
train_data = pd.read_pickle(os.path.join(data_path, 'new_sales_lag_after12.pickle')).dropna()


train_data.head()
```

```python
test_data=pd.read_csv(os.path.join(data_path, 'test.csv.gz'))
x_train_data=train_data.copy().drop(['target','item_cat_block_target_sum',
       'item_cat_block_target_mean', 'item_block_target_sum',
       'item_block_target_mean', 'shop_block_target_sum',
       'shop_block_target_mean'],axis=1)
y_train_data=train_data.target
x_test_data=test_data.copy().drop('ID',axis=1)
```

```python
x_test_data.shape
```

```python
x_test_data['date_block_num']=34
x_test_data.head()
```

```python
# merge with category id
x_test_data =pd.merge(x_test_data,items[['item_id','item_category_id']],on=['item_id'],how='left')
x_test_data.head()
```

```python
x_test_data.columns
```

```python
temp_train_data=train_data.drop('item_id',axis=1).groupby(['date_block_num','shop_id','item_category_id'],as_index=False).first()
temp_train_data.tail()
```

```python
Mforcast=34
lag_steps=[1,2,3,5,12]
new_cols=list(x_test_data.columns)
for lag in lag_steps:
    x_test_data = pd.merge(x_test_data,temp_train_data.loc[temp_train_data.date_block_num==Mforcast-lag,['shop_id','item_category_id','item_cat_block_target_sum',
       'item_cat_block_target_mean']], how='left', on=['shop_id','item_category_id']).fillna(0)
        
    x_test_data = pd.merge(x_test_data,train_data.loc[train_data.date_block_num==Mforcast-lag,['shop_id','item_id','item_block_target_sum',
       'item_block_target_mean', 'shop_block_target_sum',
       'shop_block_target_mean','target']], how='left', on=['shop_id','item_id']).fillna(0)
    
    lag_cols=['item_cat_block_target_sum_lag_'+str(lag),
       'item_cat_block_target_mean_lag_'+str(lag), 'item_block_target_sum_lag_'+str(lag),
       'item_block_target_mean_lag_'+str(lag), 'shop_block_target_sum_lag_'+str(lag),
       'shop_block_target_mean_lag_'+str(lag),'target_lag_'+str(lag)]
    new_cols=new_cols+lag_cols
    x_test_data.columns = new_cols
    
    
```

```python
x_test_data.tail()
```

```python
x_test_data.shape
```

```python
temp=x_test_data
x_test_data['December'] = temp.date_block_num.apply(lambda x: 1 if x in [23] else 0)
x_test_data['Newyear_Xmas'] = temp.date_block_num.apply(lambda x: 1 if x in [12,24] else 0)
x_test_data['Valentine_MenDay'] = temp.date_block_num.apply(lambda x: 1 if x in [13,25] else 0)
x_test_data['WomenDay'] = temp.date_block_num.apply(lambda x: 1 if x in [14,26] else 0)
x_test_data['Easter_Labor'] = temp.date_block_num.apply(lambda x: 1 if x in [15,27] else 0)
x_test_data.tail()
```
