```python
import pandas as pd
import os
import numpy as np
import pickle
#os - operating system dependent. to read/write files
#pd - pandas for data analysis/modeling

data_path = 'data/'
dataframes_path = 'dataframes/'

#get DataFrames
all_data1 = pd.read_pickle(os.path.join(dataframes_path, 'saleswithzero1.pickle'))
all_data2 = pd.read_pickle(os.path.join(dataframes_path, 'saleswithzero2.pickle'))

all_data=pd.concat([all_data1,all_data2], ignore_index=True)
all_data.tail()
```

```python
all_data.info()
```

```python
all_data.info()
```

```python
all_data.columns
```

```python
#series = Series.from_csv('daily-minimum-temperatures-in-me.csv', header=0)
lag_features=['item_cat_block_target_sum','item_cat_block_target_mean','item_block_target_sum','item_block_target_mean','shop_block_target_sum','shop_block_target_mean','item_cnt_month']
lag_steps=[1,2,3,5,12]
lag_cols=[]
temps = all_data[lag_features]
all_data = pd.concat([all_data,temps.shift(1),temps.shift(2),temps.shift(3), temps.shift(5), temps.shift(12)], axis=1)

for i in lag_steps:
    for lag_f in lag_features:
        lag_cols.append(lag_f+"_lag_"+str(i))
fixed_cols=['shop_id', 'item_id', 'date_block_num', 'date', 'item_cnt_month',
       'item_price', 'item_category_id', 'item_cat_block_target_sum',
       'item_cat_block_target_mean', 'item_block_target_sum',
       'item_block_target_mean', 'shop_block_target_sum',
       'shop_block_target_mean']  

#len(fixed_cols+lag_cols)
all_data.columns = fixed_cols+lag_cols
all_data.head()
#dataframe.columns


```

```python
all_data.shape
```

```python
all_data = all_data[all_data['date_block_num'] >= 12]
all_data.head()
```

```python
float_cols = [c for c in all_data if all_data[c].dtype == "float64"]
#int_cols =   [c for c in all_data if all_data[c].dtype == "int64"]
# Downcast
all_data[float_cols] = all_data[float_cols].astype(np.int16)
#all_data[int_cols]   = all_data[int_cols].astype(np.int32)
```

```python
all_data.info()
```

```python
temp=all_data
all_data['December'] = temp.date_block_num.apply(lambda x: 1 if x in [23] else 0).astype(bool)
all_data['Newyear_Xmas'] = temp.date_block_num.apply(lambda x: 1 if x in [12,24] else 0).astype(bool)
all_data['Valentine_MenDay'] = temp.date_block_num.apply(lambda x: 1 if x in [13,25] else 0).astype(bool)
all_data['WomenDay'] = temp.date_block_num.apply(lambda x: 1 if x in [14,26] else 0).astype(bool)
all_data['Easter_Labor'] = temp.date_block_num.apply(lambda x: 1 if x in [15,27] else 0).astype(bool)
```
